```python
import pandas as pd
import math
```

```python
df = pd.read_csv('/content/drive/MyDrive/DMML/Codebasics/titanic.csv')
df.head()
```

```python
df = df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(math.floor(df['Age'].mean()), axis='columns', inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
inputs = df.drop(['Survived'], axis='columns')
target = df['Survived']
```

```python
inputs.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
inputs['en_Sex'] = le.fit_transform(inputs['Sex'])
```

```python
inputs.head()
```

```python
inputs_n = inputs.drop(['Sex'], axis='columns')
inputs_n.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_n, target, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
ds = DecisionTreeClassifier(criterion='entropy')
```

```python
ds.fit(X_train, y_train)
```

```python
target.head()
```

```python
ds.predict([[3,35,8.0500,1]])
```

```python
ds.score(X_test, y_test)
```
