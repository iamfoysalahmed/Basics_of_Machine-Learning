Linear Regression is a Supervised Learning Technique. 
The representation of a linear equation that combines a specific set of input values (x) the solution to which is the predicted output for that set of input values (y). As such, both the input values (x) and the output value are numeric.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/MyDrive/DMML/Codebasics/canada_per_capita_income.csv')
df.head(5)
```

```python
plt.xlabel('year')
plt.ylabel('price')
plt.scatter(df.year,df.pci,color='red',marker='+')
```

```python
X = df.drop(['pci'], axis='columns')
y = df.pci
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
reg.predict([[2020]])
```
