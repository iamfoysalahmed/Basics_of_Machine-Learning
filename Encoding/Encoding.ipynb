````python
import numpy as np
import pandas as pd
import math
````

````python
df = pd.read_csv('/content/drive/MyDrive/DMML/Codebasics/weather.numeric.csv')
df.head(10)
````

````python
df.isnull().sum()
````

````python
math.floor(df['temperature'].mean())
````

````python
df1['temperature'].fillna(100, inplace=True)
df1
````

````python
df['temperature'].fillna(math.floor(df['temperature'].mean()), inplace=True) 
````
